```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
employee = pd.read_csv('Employee_HR.csv')
employee
```

Q1: For the given dataset, calculate the variance, standard deviation, and interquartile range (IQR) of the
salaries of employees in the HR and Marketing departments.





```python
hr_salary = employee[employee['Department'] == 'hr']['Salary_INR']
marketing_salary = employee[employee['Department'] == 'marketing']['Salary_INR']


```

```python
print("HR_SALARY")
hr_salary
```

```python
print("MARKETING_SALARY")
marketing_salary
```

Calculating the variance and standard deviation:

```python
variance_hr = np.var(hr_salary, ddof=0)
std_deviation_hr = np.std(hr_salary, ddof=0)
```

```python
variance_marketing = np.var(marketing_salary, ddof=0)
std_deviation_marketing = np.std(marketing_salary, ddof=0)
```

Calculating the IQR:
IQR =Q3 -Q1

```python
q1_hr = np.percentile(hr_salary, 25)
q3_hr = np.percentile(hr_salary, 75)
q1_marketing = np.percentile(marketing_salary, 25)
q3_marketing = np.percentile(marketing_salary, 75)
iqr_hr = q3_hr - q1_hr
iqr_marketing = q3_marketing - q1_marketing
```

```python
print("Variance of HR department                  : ",variance_hr )
print("Standard Deviation of HR department        : ",std_deviation_hr )
print("Variance of Marketing department           : ",variance_marketing )
print("Standard Deviation of marketing department : ",std_deviation_marketing )
print("IQR of HR department                       : ",iqr_hr )
print("IQR of Marketing department                : ",iqr_marketing )
```

Q2: What is the range of experience in the IT department?


```python
exp_it = employee[employee['Department'] == 'IT']['time_spent_company']
range_exp_it = np.ptp(exp_it)
print("The range of experience in the IT department : ", range_exp_it)
```

Q3: Create a boxplot to display the experience of each department.


```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Department',y='time_spent_company', data = employee)
plt.title('Boxplot of Experience by Department')
plt.xlabel('Department')
plt.ylabel('Time Spent (years)')
plt.show()
```
